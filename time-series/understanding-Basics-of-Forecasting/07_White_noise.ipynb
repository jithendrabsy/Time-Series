- White noise: all variables are independent and identically distributed with zero mean.
- all variables have same variance
- each value has a zero correlation with all other values in the series
- if variables are drawn from Gaussian distribution - Gaussian white noise


- white noise -> zero mean, constant variance, zero correlation

- if TS is white noise -> unpredictable
- series of errors from a model should ideally be white noise
- y(t) = signal(t)
- good model -> results in white noise

- create a line plot -> check gross features like mean, variance and obvious relationship between lagged variables
- calculate summary statistics -> mean and variance of whole series vs mean and variance of meaningful contiguous blocks in series (days, months or years)
- create ACF plot -> gross correlation between lagged values

```python
import random
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
```

```python
random.seed(12)
series = [random.gauss(0.0, 1.0) for i in range(1000)]
series = pd.Series(series)
```

```python
series.describe()
```

```python
series.plot()
plt.show()
```

```python
series.hist()
```

```python
autocorrelation_plot(series)
plt.show()
```
