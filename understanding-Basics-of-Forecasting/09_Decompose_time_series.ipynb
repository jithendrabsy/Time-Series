Time series components:
- 3 systematic components -> level, trend and seasonality
- 1 non-systematic component -> noise

- level -> the average value in the series
- trend -> increasing or decreasing value in the series
- seasonality -> repeating short-term cycle in the series
- noise -> random fluctuation in the series


- all series have level and noise
- trend and seasonality are optional

Additive model:
- y(t) = level + trend + seasonality + noise
- linear trend - straight line
- linear seasonality - same freq and amplitude


Multiplicative model:
- y(t) = level * trend * seasonality * noise
- non-linear
- nonlinear trend - curved line
- nonlinear seasonality - increasing/ decreasing amplitude and freq over time

```python
# additive decomposition
import pandas as pd
import matplotlib.pyplot as plt
import random
from statsmodels.tsa.seasonal import seasonal_decompose

series = [i+random.randrange(10) for i in range(1,100)]
result = seasonal_decompose(series, model='additive', period=1)
result.plot()
plt.show()
```

- good example where classical automatic decomposition was not able to seperate noise added from linear trend!
- more advanced decompositions available -> seasonal and trend decomposition using Loess or STL decomposition

```python
series = [i**2.0 for i in range(1,100)]
result = seasonal_decompose(series, model='multiplicative', period=1)
result.plot()
plt.show()
```

```python
# air line passengers multiplicative decomposition

df = pd.read_csv("data/AirPassengers.csv", header=0, index_col=0, parse_dates=True, squeeze=True)
result = seasonal_decompose(df, model="multiplicative")
result.plot()
plt.show()
```
